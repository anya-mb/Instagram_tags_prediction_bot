```
def derive_shortcodes_from_tag_json(data):
    shortcode_list = []

    for element in data["GraphImages"]:
        shortcode_list.append(element["shortcode"])

    return shortcode_list
```

## Get username_list by short post code

```
import requests
from bs4 import BeautifulSoup
import re
import json
```

```
# it wasn't easy to parse it with bs4, so I take regexp.
# @[a-z0-9-]+ - means find smth starts with "@" and than as many letters as possible from []. 
# Space is absent, so it's sign to stop for regexp. + - means take 1 or more symbols
NICKNAME_PATTERN = re.compile("@[a-z0-9-_-]+")

# link should be smth like that B913W0SIg3-
# remove_at - remove symbol '@' from the nickname
def get_name_by_short_post_code(short_code, remove_at=True):
    p_short_link = "https://www.instagram.com/p/" + short_code
    p_page = requests.get(p_short_link)
    if p_page.status_code == 404:
        return None
        
    p_soup = BeautifulSoup(p_page.content, 'html.parser')
    p_m = NICKNAME_PATTERN.search(str(p_soup.find("meta", property="og:description")))
    if p_m is None:
        return None
        # raise Exception("Couldn't find name by short link. Plz take a look to the page a have: " + p_page.content)
    p_nickname = p_m.group(0)
    if remove_at and p_nickname[0] == "@":
        return p_nickname[1:]
    else:
        return p_nickname
```

```
shortcode_list = []

for tag_name in tag_names:

    with open(join(tag_folder_path, tag_name, tag_name+'.json'), "r") as read_file:
        data = json.load(read_file)

    shortcode_list.extend(derive_shortcodes_from_tag_json(data))
    
    print(tag_name, len(shortcode_list))
```

```
%%time

usernames_list = []
None_links = []

for short_code in shortcode_list:
    # print(short_code)
    username = get_name_by_short_post_code(short_code)

    if username is not None:
        usernames_list.append(username)
    else:
        None_links.append(short_code)

print(len(usernames_list))
```

```
def delete_not_needed_files(username_photos_dir, not_needed_photos):
    deleted_num = 0
    for filename in not_needed_photos:
        try:
            os.remove(join(username_photos_dir, filename))
            deleted_num += 1
        except OSError as e:
            pass
    print(f"Deleted {deleted_num} in {username_photos_dir}")
```

```
class Photo_info:
    def __init__(self, tags, post_time, shortcode, likes_number, photo_file_path):
        self.tags = tags
        self.post_time = post_time
        self.shortcode = shortcode
        self.likes_number = likes_number
        # self.location = location
        # self.photo_caption = photo_caption
        self.photo_file_path = photo_file_path
```

```
def leave_only_english_tags(list_of_tags):
    ENGLISH_TAGS_PATTERN = re.compile("[^a-zA-Z0-9_]+")
    return [tag for tag in list_of_tags if ENGLISH_TAGS_PATTERN.search(tag) is None]

# testing performance
list_of_tags = ["bendenbi_rkare", "fotografdukkanim", "fotoğrafsevenler", "doğadayız", "pewpew11"]
leave_only_english_tags(list_of_tags)
```

```
%%time
# 14 tags

download = True
removed_usernames = 0
n_users = len(usernames_list)
deleted_usernames = []
not_downloaded = []

for n, name in enumerate(usernames_list):
    if (name in checked_username) | (name in deleted_usernames):
        continue

    start_time = time.time()
    print(name, f"{n+1}/{n_users}")

    if download:
        # downloading photos
        output = subprocess.run(['instagram-scraper', name, '-d', f'{photos_path}/{name}_photo/', 
                                '-t', 'image', '-m', '1000', '--media-metadata', '--include-location'])
        print(output)
        try:
            l = len(listdir(f'{photos_path}/{name}_photo/'))
            print(f"Downloading took time {time.time()-start_time}")
        except OSError as e:
            print(f"{name} not downloaded!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
            not_downloaded.append(name)
            continue

    # deriving information about photo
    info_json_path = f'{photos_path}/{name}_photo/{name}.json'
    username_photos_dir = f'{photos_path}/{name}_photo/'

    try:
        with open(info_json_path, "r") as read_file:
            data = json.load(read_file)
    except OSError as e:
        all_filemanes_in_dir = listdir(username_photos_dir)
        delete_not_needed_files(username_photos_dir, all_filemanes_in_dir)
        shutil.rmtree(username_photos_dir, ignore_errors=True)
        removed_usernames += 1
        deleted_usernames.append(name)
        print(f"Username {name} has no json")
        continue

    
    user_photo_names = set([filename for filename in listdir(username_photos_dir) if filename.endswith(".jpg")])
    
    if len(user_photo_names) <= 1:
        # os.rmdir(username_photos_dir)
        shutil.rmtree(username_photos_dir, ignore_errors=True)
        removed_usernames += 1
        deleted_usernames.append(name)
        print(f"Username {name} has no photos")
        continue

    added_photo_number = 0
    not_found_photo_number = 0
    no_tag_number = 0
    tags_num = 0
    tags_not_deleted = 0

    found_files_set = set()

    tags_list = []
    post_time_list = []
    shortcode_list = []
    likes_number_list = []
    # location_list = []
    # photo_caption_list = []
    photo_file_path_list = []

    for n_element, element in enumerate(data["GraphImages"]):

        display_url = element["display_url"] 
        photo_file_path = display_url[:display_url.find(".jpg")+4].split("/")[-1]

        try:
            tags = element["tags"]
            tags_num += len(tags)
            tags = leave_only_english_tags(tags)
            tags_not_deleted += len(tags)

        except:
            tags = []
            no_tag_number +=1

        if (photo_file_path in user_photo_names) & (len(tags) >= min_tag_number):

            added_photo_number +=1
            found_files_set.add(photo_file_path)

            post_time = element["taken_at_timestamp"]
            shortcode = element["shortcode"]
            likes_number = element["edge_media_preview_like"]["count"]
            # location = element["location"]

            # try:
            #     photo_caption = element["edge_media_to_caption"]["edges"][0]["node"]["text"]
            # except KeyError:
            #     photo_caption = 'No_caption'

            tags_list.append(tags)
            post_time_list.append(post_time)
            shortcode_list.append(shortcode)
            likes_number_list.append(likes_number)
            # location_list.append(location)
            # photo_caption_list.append(photo_caption)
            photo_file_path_list.append(f'{photos_path}/{name}_photo/{photo_file_path}')
        
        else:
            not_found_photo_number += 1

    downloaded_photos_to_delete = user_photo_names - found_files_set
    delete_not_needed_files(username_photos_dir, downloaded_photos_to_delete)
    
    print(f"For {name} added photos {added_photo_number}, not found {not_found_photo_number}, to delete {len(downloaded_photos_to_delete)}, without min tags {no_tag_number}, tags all {tags_num}, tags not deleted {tags_not_deleted}")
    
    if added_photo_number == 0:
        # delete json file and photo directory
        os.remove(info_json_path)
        
        all_filemanes_in_dir = listdir(username_photos_dir)
        delete_not_needed_files(username_photos_dir, all_filemanes_in_dir)

        os.rmdir(username_photos_dir)
        removed_usernames += 1
        deleted_usernames.append(name)
        print(f"Deleted {name}")
    
    else:

        # moving done username to users_jsons_path 
        shutil.move(info_json_path, users_jsons_path)

        # save complete file
        result = Photo_info(tags_list, post_time_list, shortcode_list, likes_number_list, photo_file_path_list)
        with open(f'{my_photo_info_jsons_path}/{name}.json', 'w') as f:
            json.dump(result.__dict__, f)
        print(f"Json saved for {name}")
    
    checked_username.append(name)
    print(f"Data preparation took time {time.time()-start_time}")
    print(" ")
```
